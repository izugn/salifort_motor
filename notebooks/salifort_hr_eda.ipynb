```python
# Create a helper fuction to calculate number of rows containing outliers in each column.

# Use 'tabulate' library for nicer results.
import tabulate as tab

def count_outliers(df, columns):
    """
    Count the number of outliers in each column of a DataFrame.

    Parameters taken:
    1. df (pd.DataFrame): The DataFrame to count outliers in.
    2. columns (list): A list of column names to count outliers in.

    Returns:
    str: A table showing the number of outliers in each column.
    """
    # Create an empty dictionary to store the results
    outliers_dict = {}

    # Loop through each column in the list
    for column in columns:
        # Calculate the 25th and 75th percentiles
        q1 = df[column].quantile(0.25)
        q3 = df[column].quantile(0.75)
        # Calculate the interquartile range (IQR)
        iqr = q3 - q1
        # Calculate the lower and upper bounds
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        # Count the number of outliers
        outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)].shape[0]
        # Add the results to the dictionary
        outliers_dict[column] = outliers

    # Create a table from the dictionary using 'tabulate' library
    table = tab.tabulate(outliers_dict.items(), headers=['Column', 'rows_with_outliers'], tablefmt='simple')

    return table
# Define all numeric columns to examine:
numeric_columns = ['satisfaction_level', 'last_evaluation', 'number_project',
       'average_monthly_hours', 'time_spend_company', 'work_accident', 'left',
       'promotion_last_5years']

# Call the helper function and set the parameters
outliers_table = count_outliers(df1, numeric_columns)
print(outliers_table)
```

```python
# Boxplot for 'time_spend_company':
plt.figure(figsize=(10, 6))
sns.boxplot(data=df1, x='time_spend_company', color='#619FCA')
plt.title('Box plot of Time Spent at the Company')
plt.xlabel('Time Spent at the Company')
plt.show()
```

```python
# Calculate size of each department, in % related to the entire dataset:
department_size = ((df1['department'].value_counts(normalize=True) / df1.value_counts(normalize=True).sum())* 100).round(2)
department_size_df = department_size.reset_index()
department_size_df.columns = ['Department', 'Size in company %']

# Calculate the turnover rate (employees who left, %) by department
turnover_rate = (df1.groupby('department')['left'].mean().sort_values(ascending=False)* 100).round(2)
turnover_rate_df = turnover_rate.reset_index()
turnover_rate_df.columns = ['Department', 'Turnover rate %']

# merge the two tables ON department
department_table = pd.merge(department_size_df, turnover_rate_df, on='Department')

department_table = department_table.sort_values(by='Turnover rate %', ascending=False)
# display the table
department_table = tab.tabulate(department_table, headers='keys', tablefmt='simple', showindex=False)
print("Department size and turnover rate:\n", department_table)
```

```python
# Calculate the order of departments by the number of employees who left
sns.histplot(data=df1, x='department', hue='left', discrete=1, multiple='stack', shrink=.6, alpha=0.7)
plt.xticks(rotation=45)
plt.title('Department turnover')
plt.legend(title='Turnover', labels=['Left', 'Stayed'])
```

```python
# Create 2 subsets of df1: one for employees who left and one for employees who stayed:
df_left = df1[df1['left'] == 1]
df_stayed = df1[df1['left'] == 0]

# Combine the two dataframes for boxplots
df_combined = pd.concat([df_stayed, df_left])
```

```python
# Boxplot mean satisfaction_level by department (df stayed and df left) on the same plot
fig, ax = plt.subplots(figsize=(16, 6))
sns.boxplot(data=df_combined, x='department', y='satisfaction_level', hue='left', palette=['#619FCA', '#FFA556'], width=0.5)
ax.set_title('Mean Satisfaction Level by Department')
ax.set_xlabel('Department')
ax.set_ylabel('Satisfaction Level')
plt.tight_layout()
ax.tick_params(axis='x', rotation=45)
plt.legend(title='Turnover', loc='upper right')
plt.show()
```

```python
# Boxplot for Time at Company Analysis by Satisfaction level
plt.figure(figsize=(16, 6))
sns.boxplot(data = df1, x='time_spend_company', y='satisfaction_level', hue='left', palette=['#619FCA', '#FFA556'], width=0.5)
plt.title('Satisfaction Level by Years at Company and Turnover Status')
plt.xlabel('Years spent at company')
plt.ylabel('Satisfaction level')
plt.legend(title = 'Turnover')
plt.show()
```

```python
# Calculate the turnover rate by yesr of service
service_turnover = (df1.groupby('time_spend_company')['left'].mean()* 100).round(2)
service_turnover = service_turnover.reset_index()
service_turnover.columns = ['Years of service', 'Turnover rate']
service_turnover = tab.tabulate(service_turnover, headers='keys', tablefmt='simple', showindex=False)
print("Turnover rate by years of service:\n", service_turnover)
```
